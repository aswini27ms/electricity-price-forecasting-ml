```python
df['time'] = pd.to_datetime(df['time'], utc=True) # Ensure 'time' column is datetime and handle timezones
df['hour'] = df['time'].dt.hour        # 0–23
df['day'] = df['time'].dt.day          # 1–31
df['month'] = df['time'].dt.month      # 1–12
df['dayofweek'] = df['time'].dt.dayofweek   # 0=Mon, 6=Sun
```

```python
df.drop('time', axis=1, inplace=True)
```

```python
df.drop([
    'generation fossil coal-derived gas',
    'generation fossil oil shale',
    'generation fossil peat',
    'generation geothermal',
    'generation marine',
    'generation wind offshore'
], axis=1, inplace=True)
```

```python
import warnings
warnings.filterwarnings('ignore')
for i in df.select_dtypes(include="number").columns:
  sns.histplot(data=df,x=i)
  plt.show()
```

```python
# For outliers use capping (Winsorization)
for col in df.select_dtypes(include="number").columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)
```

```python
#correlation with heatmap to interpret the relation and multicollinearity
s = df.select_dtypes(include = "number").corr()
plt.figure(figsize=(30,30))
sns.heatmap(s,annot = True)
```

```python
plt.figure(figsize=(8,10))
price_corr = df.corr()[['price actual']].sort_values(by='price actual', ascending=False)
sns.heatmap(price_corr, annot=True, cmap='coolwarm')
plt.show()
```
